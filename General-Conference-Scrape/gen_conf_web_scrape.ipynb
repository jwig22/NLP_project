##### https://www.churchofjesuschrist.org/study/general-conference/2022/10

## This next bit is how to find the text from the site

#### HTML Block is: div class="body-block"

```python
from requests_html import HTMLSession
from bs4 import BeautifulSoup
import requests
import os
from urllib.parse import urlparse

URL = "https://www.churchofjesuschrist.org/study/general-conference/2022/10/"
```

## Next is to figure out how to iterate over all of the sessions and talks

```python
reqs = requests.get(URL)

soup = BeautifulSoup(reqs.content, 'html.parser')
 

urls = []

for link in soup.find_all('a'):
    urls.append("https://www.churchofjesuschrist.org" + link.get('href'))
# urls
```

```python
# substring1 = "session"
substring2 = "lang=eng"
# urls2 = [item for item in urls if substring1 not in item]
urls_new = [item for item in urls if substring2 not in item]

if urls_new[0] == urls_new[1]:
    print("duplicate")
    urls_new.pop(0)
else:
    print("No duplicate found")
# urls_new
```

```python
print(urls_new)
```

## Try to iterate over gathered URLs and append text to a list

```python
sel = 'body > div > div > div > main > div > div > div > section > div > article > div > div' 

all_talks = []
for link in urls_new:
    try:
        session = HTMLSession()
        r = session.get(link)
        talk = r.html.find(sel, first=True).text
        if talk == "Record Your Impressions\nCreate a Note":
            talk = ""
    except:
        print("URL did not have a talk: ", link)
        talk = ""
    all_talks.append(os.path.basename(link))
    all_talks.append("\n")
    all_talks.append(talk)
    all_talks.append("\n\n")

    
```

```python
with open("gen_conf_oct_22.txt","w+") as f:
        for talk in all_talks:
            f.write(talk)
```
